```python
import pandas as pd

df = pd.read_csv('./sample_data/services_dataset.csv')

display(df.head())
display(df.info())
```

```python
display(df.isnull().sum())
```

```python
df_encoded = pd.get_dummies(df, columns=['service_name'], drop_first=True)

X = df_encoded[['estimated_units'] + [col for col in df_encoded.columns if 'service_name_' in col]]
y_avg = df_encoded['total_price']
y_min = df_encoded['total_price'] # Using total_price as a placeholder
y_max = df_encoded['total_price'] # Using total_price as a placeholder

display(X.head())
display(y_avg.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_avg_train, y_avg_test, y_min_train, y_min_test, y_max_train, y_max_test = train_test_split(
    X, y_avg, y_min, y_max, test_size=0.2, random_state=42
)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_avg_train:", y_avg_train.shape)
print("Shape of y_avg_test:", y_avg_test.shape)
print("Shape of y_min_train:", y_min_train.shape)
print("Shape of y_min_test:", y_min_test.shape)
print("Shape of y_max_train:", y_max_train.shape)
print("Shape of y_max_test:", y_max_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

model_avg = RandomForestRegressor(random_state=42)

model_avg.fit(X_train, y_avg_train)

print("Average Model trained successfully.")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_avg_pred = model_avg.predict(X_test)
```

```python
mse_avg = mean_squared_error(y_avg_test, y_avg_pred)
r2_avg = r2_score(y_avg_test, y_avg_pred)

print(f"Average Model - MSE: {mse_avg:.2f}, R2: {r2_avg:.2f}")
```

### Etapa 1: Codificar todas as variáveis categóricas relevantes e definir as features (X)

```python
categorical_cols = ['category', 'charging_unit', 'complexity_level', 'urgency', 'service_name']
df_encoded_full = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Definir X (features) incluindo todas as colunas numéricas originais e as novas colunas one-hot encoded
# Excluímos 'total_price' que é a nossa variável alvo
X_new = df_encoded_full.drop(columns=['total_price'])
y_avg_new = df_encoded_full['total_price']
y_min_new = df_encoded_full['total_price'] # Placeholder
y_max_new = df_encoded_full['total_price'] # Placeholder

display(X_new.head())
print(f"New X shape: {X_new.shape}")
```

### Etapa 2: Dividir os dados novamente com as novas features

```python
from sklearn.model_selection import train_test_split

X_train_new, X_test_new, y_avg_train_new, y_avg_test_new, y_min_train_new, y_min_test_new, y_max_train_new, y_max_test_new = train_test_split(
    X_new, y_avg_new, y_min_new, y_max_new, test_size=0.2, random_state=42
)

print("Shape of X_train_new:", X_train_new.shape)
print("Shape of X_test_new:", X_test_new.shape)
print("Shape of y_avg_train_new:", y_avg_train_new.shape)
print("Shape of y_avg_test_new:", y_avg_test_new.shape)
```

### Etapa 3: Treinar o modelo RandomForestRegressor com as novas features

```python
from sklearn.ensemble import RandomForestRegressor

model_avg_new = RandomForestRegressor(random_state=42)
model_avg_new.fit(X_train_new, y_avg_train_new)

print("Novo modelo de média treinado com sucesso.")
```

### Etapa 4: Avaliar o novo modelo

```python
from sklearn.metrics import mean_squared_error, r2_score

y_avg_pred_new = model_avg_new.predict(X_test_new)

mse_avg_new = mean_squared_error(y_avg_test_new, y_avg_pred_new)
r2_avg_new = r2_score(y_avg_test_new, y_avg_pred_new)

print(f"Novo Modelo de Média - MSE: {mse_avg_new:.2f}, R2: {r2_avg_new:.2f}")
```
